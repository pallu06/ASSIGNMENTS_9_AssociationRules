```python
import pandas as pd
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
# conda install -c conda-forge mlxtend
```

```python
book = pd.read_csv("book.csv")
```

```python
book
```

#We can do preprocessing if we have catagorical values

# Apriori Algorithm

```python
frequent_itemsets = apriori(book, min_support=0.5, use_colnames=True)
frequent_itemsets
```

```python
#finding the rules for different support values
frequent_itemsets = apriori(book, min_support=0.6, use_colnames=True)
frequent_itemsets
```

```python
frequent_itemsets = apriori(book, min_support=0.2, use_colnames=True)
frequent_itemsets
```

```python
frequent_itemsets = apriori(book, min_support=0.4, use_colnames=True)
frequent_itemsets
```

```python
frequent_itemsets = apriori(book, min_support=0.1, use_colnames=True)
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=0.7)
rules
```

```python
import matplotlib.pyplot as plt
plt.scatter(rules.support,rules.confidence)
```

```python
#A high conviction value means that the consequent is highly depending on the antecedent and range [0 inf]
```

```python
rules.sort_values('lift',ascending = False)[0:20]
```

```python
rules[rules.lift>1]
```
